```python
from googletrans import Translator

#日本
def BackTranslationJa(text):
    translator = Translator()
    return translator.translate(translator.translate(text, dest="ja", src='en').text, dest="en", src="ja").text

#フランス
def BackTranslationFr(text):
    translator = Translator()
    return translator.translate(translator.translate(text, dest="fr", src='en').text, dest="en", src="fr").text

#ドイツ
def BackTranslationGe(text):
    translator = Translator()
    return translator.translate(translator.translate(text, dest="de", src='en').text, dest="en", src="de").text

#スペイン
def BackTranslationSp(text):
    translator = Translator()
    return translator.translate(translator.translate(text, dest="es", src='en').text, dest="en", src="es").text

#オランダ
def BackTranslationNl(text):
    translator = Translator()
    return translator.translate(translator.translate(text, dest="nl", src='en').text, dest="en", src="nl").text
```
