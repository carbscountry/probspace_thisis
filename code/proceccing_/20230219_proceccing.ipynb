## 形態解析解析


```python
# 単語出現頻度
def remove_stopword(x):
    stopwords_list = set(stopwords.words("english") + list(string.punctuation))
    return [y for y in x if y not in stopwords_list]


def clean_token(_text):
    # トークン化、ストップワードの除去
    text_tokens = nltk.word_tokenize(re.sub(r'[^a-zA-Z0-9_-]',' ',_text.lower())) #-は残し、tokenした後消す
    stopwords_list = set(stopwords.words("english") + list(string.punctuation))
    text_tokens =  [word for word in text_tokens if not word in stopwords_list]
    return [re.sub("[^a-zA-Z0-9_]"," ",word) for word in text_tokens ]



```

```python
train_df = train_df.with_column(
    [
    (pl.col("keywords").apply(lambda x:x.lower().replace('-',' ').split(',')).alias("token_keyword")),
    (pl.col("abstract").apply(lambda x:clean_token(x)).alias("token_abstract"))
    ]
)
```

```python
test_df = test_df.with_column(
    [
    (pl.col("keywords").apply(lambda x:x.lower().replace('-',' ').split(',')).alias("token_keyword")),
    (pl.col("abstract").apply(lambda x:clean_token(x)).alias("token_abstract"))
    ]
)
```
